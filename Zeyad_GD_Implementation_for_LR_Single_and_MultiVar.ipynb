## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with zeros. Also choose the Learning rate.

```python
theta_0 = 0
theta_1 = 0
alpha = 0.0001
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.

```python
h_x = theta_0 + theta_1 * X
h_x
```

### Step3: Calculate the MSE Cost function ùë±(theta_0,theta_1 ).
<b>Note:</b> use norm square of the error vector.

```python
error = h_x - y
m = len(y)
mse= (1/(2*m))* np.sum(error**2)
print("Error vector = \n",error ,"\n \n J = ",mse)
```

```python
d_theta_0 = (1/m)*np.sum(error)
d_theta_0
```

```python
d_theta_1  = (1/m)*np.sum(error*X)
d_theta_1
```

### Step5: Update the parameters.

```python
theta_0 = theta_0 - d_theta_0*alpha
theta_1 = theta_1 - d_theta_1*alpha
print("theta_0_new= ",theta_0)
print("\ntheta_1_new= ",theta_1)
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.
#### The objective from this step is to combine all the previous steps and iterate untill you either achieve the maximum number of iterations or reach the minimum.

```python
theta_0 = 0
theta_1 = 0
alpha = 0.01
mse_old=0
for i in range(500):
  h_x = theta_0 + theta_1 * X
  error = h_x - y
  mse= (1/(2*m))* np.sum(error**2)
  if abs(mse_old-mse)< 0.0001:
    print(f'****************** Training Report ********************\n')
    print(f'Gradient Descent converged after {i} iterations\n')
    print(f'theta_0_Opt : {theta_0}')
    print(f'theta_1_Opt : {theta_1}\n')
    print(f'Error Vector:\n{error}\n')
    print(f'Cost = {mse}\n')
    print(f'h(x) = y_predict:\n{h_x}\n')
    print(f'y_actual:\n{y}\n')
    break
  else:
    d_theta_0 = (1/m)*np.sum(error)
    d_theta_1  = (1/m)*np.sum(error*X)
    theta_0 = theta_0 - d_theta_0*alpha
    theta_1 = theta_1 - d_theta_1*alpha
    mse_old = mse

    print("\n****************** Iteration", i, "********************\n")
    print("h(x):", h_x)
    print("\nError Vector:")
    print(error)
    print("\nj =", mse)
    print("\nGradient Vector:")
    print(d_theta_0, d_theta_1)
    print("\ntheta_0_new :", theta_0)
    print("theta_1_new :", theta_1)
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x

```python
h_x = theta_0 + theta_1*X
h_x
```
